```python
data = pd.read_csv('final_project_data.csv')
```

```python
questions = ['fi1','fi2','fi3','fi4','td1','td2','td3','td4','td5','he1','he2','he3','he4','pwt1','pwt2','pwt3']
```

```python
print(data.info())
```

```python
rows_list = []
for q in questions:
    for row in data.iterrows():
        new_row = {}
        new_row['user_id'] = row[1]['user_id'];
        new_row['question'] = q
        new_row['answer'] = row[1][q]
        new_row['experiment_device'] = row[1]['experiment_device']
        new_row['Is_Filler'] = row[1]['Is_Filler']
        new_row['Is_Inter'] = row[1]['Is_Inter']
        new_row['Is_Rel'] = row[1]['Is_Rel']
        new_row['Required_to_Switch'] = row[1]['Required_to_Switch']
        new_row['gender'] = row[1]['gender']
        new_row['age_group'] = row[1]['age_group']
        new_row['education_level'] = row[1]['education_level_calc']
        new_row['device_type_calc'] = row[1]['device_type_calc']
        new_row['open_close_space'] = row[1]['open_close_space']
        new_row['private_public_space'] = row[1]['private_public_space']
        new_row['large_small_space'] = row[1]['large_small_space']
        new_row['noisy_quiet_space'] = row[1]['noisy_quiet_space']
        new_row['dark_bright_space'] = row[1]['dark_bright_space']
        new_row['crowded_space'] = row[1]['crowded_space']
        new_row['pc_hours'] = row[1]['pc_hours_calc']
        new_row['mobile_hours'] = row[1]['mobile_hours_calc']
        new_row['online_flights'] = row[1]['online_flights_calc']
        rows_list.append(new_row)

glmm_data = pd.DataFrame(rows_list)    
display(glmm_data)
```

```python
glmm_data.to_csv('data_for_glmm.csv', index = False)
```
